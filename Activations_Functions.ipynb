```python
#sigmoid activation
#It is differentiable &
#It maps the output into the range [0,1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating the sigmoid function

def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
#creating  an array with linear spaced between -10 and 10

x = np.linspace(-10,10,50)
print(x)

# finding the sigmoid values for x values
y = sigmoid(x)
#the output of this ranging between 0 to 1
print(y)
```

```python
#plotting the graph to demonstrate sigmoid functoin

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Function")
plt.grid(True)
plt.show()
```

Tanh Activation Function :


It maps the output into the range(-1,1)
It almost better than sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)


# Formula :
# tanhx = ez - e-z / ez + e-z

# Generate linearly spaced values
x = np.linspace(-10, 10, 50)

# Apply tanh function
y = tanh(x)

# this can be also directly  written as
# y= np.tanhx(x)

# Plot tanh function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('Tanh Activation Function')
plt.grid(True)
plt.show()
```

ReLU (Rectified Linear Unit) Activation Function :

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

# Generate linearly spaced values
x = np.linspace(-10, 10, 50)

# Apply ReLU function
y = relu(x)

# Plot ReLU function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Activation Function')
plt.grid(True)
plt.show()
```

Softmax Activation Function :

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exps = np.exp(x - np.max(x))  # Subtracting max value for numerical stability
    return exps / np.sum(exps)

# Generate linearly spaced values
x = np.linspace(-10, 10, 50)

# Apply softmax function
y = softmax(x)

# Plot softmax function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('softmax(x)')
plt.title('Softmax Activation Function')
plt.grid(True)
plt.show()
```
